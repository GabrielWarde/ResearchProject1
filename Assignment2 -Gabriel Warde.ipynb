```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.metrics import adjusted_rand_score

def k_means(data, k, max_iters=100):
    centroids = data[np.random.choice(range(len(data)), k, replace=False)]
    for _ in range(max_iters):
        # Assign each data point to the nearest centroid
        distances = np.linalg.norm(data[:, np.newaxis] - centroids, axis=2)
        labels = np.argmin(distances, axis=1)
        
        # Update centroids
        new_centroids = np.array([data[labels == i].mean(axis=0) for i in range(k)])
        
        # Check for convergence
        if np.all(centroids == new_centroids):
            break
        
        centroids = new_centroids
    
    return labels, centroids

data = load_iris()
X = data.data
y = data.target 
feature_names = data.feature_names  # Getting the feature names
k = 3  

num_iterations = 50
adjusted_rand_scores1 = []

for _ in range(num_iterations):
    custom_labels, custom_centroids = k_means(X, k)
    adjusted_rand = adjusted_rand_score(y, custom_labels)
    adjusted_rand_scores1.append(adjusted_rand)

for i, score in enumerate(adjusted_rand_scores1):
    print(f'Run {i + 1} - Adjusted Rand Score: {score:.4f}')

mean_ars = np.mean(adjusted_rand_scores1)
std_ars = np.std(adjusted_rand_scores1)
print(f'\nMean Adjusted Rand Score: {mean_ars:.4f}')
print(f'Standard Deviation of Adjusted Rand Scores: {std_ars:.4f}')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score

data = load_iris()
X = data.data
y_true = data.target

n_clusters = 3

# Initialize variables to store best clustering results
best_custom_labels = None
best_custom_ars = 0
best_sklearn_labels = None
best_sklearn_ars = 0

# Run K-Means 50 times for both implementations 
num_iterations = 50
for _ in range(num_iterations):
    # My Custom K-Means
    custom_labels, _ = k_means(X, n_clusters)
    custom_ars = adjusted_rand_score(y_true, custom_labels)
    if custom_ars > best_custom_ars:
        best_custom_ars = custom_ars
        best_custom_labels = custom_labels

    # Scikit-learn's KMeans
    sklearn_kmeans = KMeans(n_clusters=n_clusters, random_state=None).fit(X)
    sklearn_labels = sklearn_kmeans.labels_
    sklearn_ars = adjusted_rand_score(y_true, sklearn_labels)
    if sklearn_ars > best_sklearn_ars:
        best_sklearn_ars = sklearn_ars
        best_sklearn_labels = sklearn_labels

# Create scatter plots for the best solutions
def create_scatter_plot(X, labels, title):
    plt.figure()
    for i in range(n_clusters):
        plt.scatter(X[labels == i, 0], X[labels == i, 1], label=f'Cluster {i + 1}')
    plt.title(title)
    plt.legend()
    plt.show()

# Create scatter plots for the best solutions for both implementations
create_scatter_plot(X, best_custom_labels, 'Best Custom K-Means Solution')
create_scatter_plot(X, best_sklearn_labels, 'Best scikit-learn KMeans Solution')
```

```python
import numpy as np
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

# Defining a function to calculate the WCSS for a given clustering
def calculate_wcss(data, labels, centroids):
    wcss = 0
    for i in range(len(centroids)):
        cluster_data = data[labels == i]
        cluster_center = centroids[i]
        wcss += np.sum((cluster_data - cluster_center) ** 2)
    return wcss

data = load_iris()
X = data.data

n_clusters = 3 

# Initialize variables to store WCSS values
wcss_values = []

# Run K-Means 50 times and record WCSS at each iteration
num_iterations = 50
max_iters = 100
for _ in range(num_iterations):
    centroids = X[np.random.choice(range(len(X)), n_clusters, replace=False)]
    wcss_iterations = []

    for _ in range(max_iters):
        distances = np.linalg.norm(X[:, np.newaxis] - centroids, axis=2)
        labels = np.argmin(distances, axis=1)

        new_centroids = np.array([X[labels == i].mean(axis=0) for i in range(n_clusters)])

        wcss = calculate_wcss(X, labels, new_centroids)
        wcss_iterations.append(wcss)

        if np.all(centroids == new_centroids):
            break

        centroids = new_centroids

    wcss_values.append(wcss_iterations)

# Convergence plot
plt.figure(figsize=(15, 15))
for i in range(num_iterations):
    plt.plot(range(len(wcss_values[i])), wcss_values[i], label=f'Run {i + 1}')

plt.xlabel('Number of Iterations')
plt.ylabel('WCSS (Within-Cluster Sum of Squares)')
plt.title('K-Means* Convergence Plot (50 Runs)')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score
from sklearn.datasets import load_digits

#loading dataset
data, labels = load_digits(return_X_y=True)

# Initialize variables to store the clustering results
custom_ars_scores = []
sklearn_ars_scores = []

# 50 runs of K-Means using my custom implementation
num_iterations = 50
for _ in range(num_iterations):
    # Custom K-Means implementation
    custom_labels, _ = k_means(data, 10)
    custom_ars = adjusted_rand_score(labels, custom_labels)
    custom_ars_scores.append(custom_ars)

# 50 runs of K-Means scikit-learn's
for _ in range(num_iterations):
    kmeans = KMeans(n_clusters=10, random_state=None).fit(data)
    sklearn_labels = kmeans.labels_
    sklearn_ars = adjusted_rand_score(labels, sklearn_labels)
    sklearn_ars_scores.append(sklearn_ars)

# box plot comparing the accuracy performances of custom K-Means and scikit-learn's KMeans
plt.boxplot([custom_ars_scores, sklearn_ars_scores], labels=['Custom K-Means', 'scikit-learn KMeans'])
plt.ylabel('Adjusted Rand Score')
plt.title('Comparison of K-Means Implementations (50 Runs)')
plt.show()

# Convergence plots for both implementations
def k_means_convergence(data, k, max_iters=100):
    wcss_values = []
    centroids = data[np.random.choice(range(len(data)), k, replace=False)]
    for _ in range(max_iters):
        distances = np.linalg.norm(data[:, np.newaxis] - centroids, axis=2)
        labels = np.argmin(distances, axis=1)
        new_centroids = np.array([data[labels == i].mean(axis=0) for i in range(k)])
        wcss = calculate_wcss(data, labels, new_centroids)
        wcss_values.append(wcss)
        if np.all(centroids == new_centroids):
            break
        centroids = new_centroids
    return wcss_values

custom_wcss_values = []
sklearn_wcss_values = []

for _ in range(num_iterations):
    custom_wcss = k_means_convergence(data, 10)
    custom_wcss_values.append(custom_wcss)

# Creating a list to store scikit-learn's WCSS values
sklearn_wcss_values = []
for _ in range(num_iterations):
    kmeans = KMeans(n_clusters=10, random_state=None).fit(data)
    sklearn_wcss_values.append(kmeans.inertia_)

plt.figure(figsize=(12, 6))
for wcss in custom_wcss_values:
    plt.plot(range(len(wcss)), wcss, color='b', alpha=0.5)

# Plotting scikit-learn's WCSS values
for sklearn_wcss in sklearn_wcss_values:
    plt.plot(range(len(wcss)), [sklearn_wcss] * len(wcss), color='r', alpha=0.5)

plt.xlabel('Number of Iterations')
plt.ylabel('WCSS (Within-Cluster Sum of Squares)')
plt.title('K-Means Convergence Plot (50 Runs)')
plt.legend(['Custom K-Means', 'scikit-learn KMeans'])
plt.show()
```
